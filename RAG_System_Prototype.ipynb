```python
import chromadb

client = chromadb.Client()

collection_name = 'product_info'
collection = client.create_collection(name=collection_name)

ids = [f"doc_{i}" for i in range(len(database))]
embeddings = [model.encode(text).tolist() for text in database]

collection.add(ids=ids, embeddings=embeddings, documents=database)

print("Data successfully stored in Chroma database.")
```

```python
query = "Give me some info on Audio & Video Connectors"
query_embedding = model.encode(query).tolist()
results = collection.query(query_embeddings=[query_embedding], n_results=1)

matching_id = results['ids'][0][0]
matching_text = results['documents'][0][0]
print(f"Text similar to the query: {matching_text}")
```
